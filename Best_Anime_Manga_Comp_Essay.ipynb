# Results

To help answer our question we created 3 visualizations, a WordCloud, a Scatter Plot, and a Horizontal Bar Graph:

    The most important graph we made was the scatterplot, which made each anime on both lists a point. It compares the average user rating to the rank on the list (the lower the number, the higher the rank), allowing us to see where each anime falls in ratings across both lists. We were also able to analyze the scatterplot and return the titles and information of each point, as seen below:

```python
#MAKES THE DATA
data_fav = Visualizations.make_data_set("favorite_anime.csv")
data_pop = Visualizations.make_data_set("popular_anime.csv")

#COMBINES THE DATA
data = Visualizations.combine_data_sets(data_fav, data_pop)

#VISUALIZATION 1
Visualizations.make_scatter_plot(data)

print("The Highest Rated and Ranked is:")
print(Visualizations.analyze_scatter_plot_max(data))
print("The Lowest Rated and Ranked is:")
print(Visualizations.analyze_scatter_plot_min(data))
```

From this, we observed that the highest rated and ranked anime across both lists was Fullmetal Alchemist: Brotherhood and the lowest was Sword Art Online II.

    We also created a word cloud with all the titles from both lists. We wanted to see if the most mentioned anime titles could be the highest rated/ranked.
